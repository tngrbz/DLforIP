```python
import os
import sys
import subprocess
```

```python
def remove_zone_identifier(folder_path):
    # Walk through the directory
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            zone_identifier_path = file_path + ":Zone.Identifier"
            
            # Check if the Zone.Identifier stream exists
            if os.path.exists(zone_identifier_path):
                try:
                    # Remove the Zone.Identifier stream using a Bash command
                    subprocess.run(['rm', '-f', f"{zone_identifier_path}"], check=True)
                    print(f"Removed Zone.Identifier from {file_path}")
                except subprocess.CalledProcessError as e:
                    print(f"Failed to remove Zone.Identifier from {file_path}: {e}")

# Set the path to the directory where you want to remove Zone.Identifier streams
folder_path = './data/'
remove_zone_identifier(folder_path)
```
